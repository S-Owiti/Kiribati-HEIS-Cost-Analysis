```python
#Setting up a clustered column chart for comparison
bar_width = 0.2
r1 = np.arange(len(dataset['Unnamed: 0']))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
r5 = [x + bar_width for x in r4]

bar_width = 0.2
categories = dataset['Unnamed: 0']
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

#creating clustered column chart
plt.figure(figsize = (10, 6))

#plotting bars
plt.bar(r1, dataset['S. Tarawa'], color = 'b', width = bar_width, edgecolor = 'grey', label = 'Average Annual Household Income')
plt.bar(r2, dataset['Nothern'], color = 'g', width = bar_width, edgecolor = 'grey', label = 'Average Annual Household Expenditure')
#plt.bar(r2, dataset['Central'], color = 'g', width = bar_width, edgecolor = 'grey', label = 'Average Annual Household Expenditure')

#adding x-ticks
plt.xlabel('Residence', fontweight = 'bold')
plt.xticks([r + bar_width for r in range(len(dataset['Unnamed: 0']))], dataset['Unnamed: 0'])

#adding labels and titles
plt.ylabel('Values', fontweight = 'bold')
#plt.yticks([r + bar_width for r in range(len(dataset['Unnamed: 0']))], dataset['Unnamed: 0'])
plt.title('Clustered Column Chart', fontweight = 'bold')

#adding legends
plt.legend()

#display plot
plt.show()
```
